```python
cnt = 0
for index,group in df.groupby(["materialnumber_marc"]):
    plt.figure(figsize=(16,8))
    print(index)
    for index2,group2 in group.groupby(["movementtype"]):
#         print(index2)  
        print()
        plt.plot( group2.set_index(["postingdate"])["quantity_sum"], marker='.', label=index2)
        plt.title(index)
    plt.xlabel("time")
    plt.ylabel('quantity')
    plt.legend()
#     plt.show()
    plt.savefig("/home/bharat/proj/hackathon/graphs/C004/new/" + 'Graph_{}.png'.format(index), format="PNG")  
```

```python
cnt = 0
for index,group in df.groupby(["materialnumber_marc","movementtype"]):
    if(index[1] == 551):
        cnt+=1
        print(cnt)
        print(index)
#         print(index[0],index[1])
        n_df1 = df.groupby(["materialnumber_marc","movementtype"]).get_group((index[0],index[1]))
        n_df1.postingdate = n_df1.postingdate.apply(lambda x : pd.to_datetime(x,format = '%Y%m%d', errors='ignore'))
        n_df1 = n_df1.sort_values('postingdate')    
        del n_df1["plant"]
        del n_df1["materialnumber_marc"]
        del n_df1["movementtype"]
        n_df1 = n_df1.sort_values('postingdate')
        n_df1.set_index('postingdate',inplace=True)
        n_df1.index = pd.to_datetime(n_df1.index, unit='s')
#         print(n_df1)
        plt.figure(figsize=(16,8))
        plt.plot( n_df1, marker='.', color='blue')
        plt.xlabel("time")
        plt.ylabel('quantity')
#         plt.show()        
#         fig.set_size_inches(10,5)
        plt.savefig("/home/bharat/proj/hackathon/graphs/rC004/551/" + 'Graph_{}.png'.format(index), format="PNG")  
    
        plt.close()
        if(cnt == 200): 
            break
# # print(md)
```
